# ---------------------------------------------------------------------------------------------------------------

# Handeling Price Range Column
    - Price range column contain 4 unique values which are [1,2,3,4]
    - Replacing these values with 
        1 ---> Inexpensive
        2 ---> Moderately Expensive
        3 ---> Expensive
        4 ---> Very Expensive

```python
DF1['Price range'].unique()
```

```python
def handlePR(value):
    if value == 4:
        return 'Very Expensive'
    elif value == 3:
        return 'Expensive'
    elif value == 2:
        return 'Moderately Expensive'
    else:
        return 'Inexpensive'
```

```python
DF1['Price range'] = DF1['Price range'].apply(handlePR)
```

```python
DF1.head(3)
```

# ---------------------------------------------------------------------------------------------------------------

# Handling column "Avg Cost For Two"
    - In this column contain numeric values so the type should be integer.
    - But Comma Style Formatting and '-' are forcing to show as object.
    - Removing the comma and '-' will solve the problem and promote it as integer value.

```python
def handleCOST(value):
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    elif '-' in value:
        value = value.replace('-',str(np.nan)) 
        return float(value)
    else:
        return float(value)
```

```python
DF1['Avg Cost for two'] = DF1['Avg Cost for two'].apply(handleCOST)
```

```python
DF1['Avg Cost for two'].isnull().sum()
```

### Removing '-' creates null values 
    Will fill these null values with average of the column.

```python
DF1['Avg Cost for two'].fillna(DF1['Avg Cost for two'].mean(), inplace=True)
DF1['Avg Cost for two'] = DF1['Avg Cost for two'].round(decimals = 0)
```

```python
DF1.info()
```

# ---------------------------------------------------------------------------------------------------------------

### Creating A New Column To Represent Avg. Cost in USD.

```python
DF1['Currency'].unique()
```

```python
usd = {'Dollar($)': 1.0,'Botswana Pula(P)': 0.079 ,'Brazilian Real(R$)': 0.20,'Emirati Diram(AED)': 0.27,'Indian Rupees(Rs.)':0.012,
       'Indonesian Rupiah(IDR)': 0.000067, 'NewZealand($)':0.65, 'Pounds(\x8c£)':1.24,
       'Qatari Rial(QR)': 0.27, 'Rand(R)':0.058, 'Sri Lankan Rupee(LKR)':0.0028,
       'Turkish Lira(TL)':0.053}
usd
```

```python
DF1['Cost in USD'] = DF1.apply(lambda x: x['Avg Cost for two']*usd[x['Currency']], axis=1).round(2)
```

# Handeling Cuisines Column
    - Column contains more then one cuisines
    - Taking out first as the speciality of the restaurant

```python
Cue = DF1['Cuisines'].str.split(",", n = 1, expand = True)
Cue.drop(1,axis = 1, inplace = True)
Cue.reset_index(inplace= True)
Cue.head()
```

```python
DF1 = DF1.merge(Cue, how = 'left',on = 'index')
DF1.rename(columns = {0 : 'Speciality'}, inplace = True)
DF1.head(5)
```

```python
DF = DF1.merge(DF2, how = 'left',on = 'Country Code')
DF.head(5)
```
